## Data Wrangling

### Gathering Data

### 1. Memuat tabel day_df

```python
day_df = pd.read_csv("data/day.csv")
day_df.head()
```

##### Mengubah judul kolom agar lebih mudah dipahami serta mengubah data berbentuk angka menjadi keterangan

```python
day_df.rename(columns={'dteday':'datetime',
                        'weathersit':'weather_condition',
                        'hum':'humidity',
                        'mnth':'month',
                        'cnt':'count',
                        'yr':'year'},inplace=True)
```

```python
day_df['weather_condition']= day_df['weather_condition'].map({1: 'Clear to partly cloudy', 2: 'Misty cloudy to misty',
                                                              3: 'Light snow to light rain', 4: 'Heavy rain to snow'})
day_df['season']= day_df['season'].map({1:'Springer', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
day_df['year'] = day_df['year'].apply(lambda x: '2011' if x == 0 else('2012' if x == 1 else 'error'))
```

##### Memperbaiki tipe data

```python
# Mengubah tipe data datetime 
day_df['datetime'] = pd.to_datetime(day_df.datetime)

# Tipe data kategorikal
day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_condition'] = day_df.weather_condition.astype('category')

day_df.head()
```

#### 2. Memuat tabel hour.df

```python
hour_df = pd.read_csv("data/hour.csv")
hour_df.head()
```

##### Mengubah judul kolom agar lebih mudah dipahami serta mengubah data berbentuk angka menjadi keterangan

```python
hour_df.rename(columns={'dteday':'datetime',
                        'weathersit':'weather_condition',
                        'hum':'humidity',
                        'mnth':'month',
                        'cnt':'count',
                        'yr':'year',
                        'hr':'hour'},inplace=True)
```

```python
hour_df['weather_condition']= hour_df['weather_condition'].map({1: 'Clear to partly cloudy', 2: 'Misty cloudy to misty',
                                                              3: 'Light snow to light rain', 4: 'Heavy rain to snow'})
hour_df['season']= hour_df['season'].map({1:'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
hour_df['year'] = hour_df['year'].apply(lambda x: '2011' if x == 0 else('2012' if x == 1 else 'error'))
```

##### Memperbaiki tipe data

```python
# Mengubah tipe data datetime 
hour_df['datetime'] = pd.to_datetime(hour_df.datetime)

# Tipe data kategorikal
hour_df['season'] = hour_df.season.astype('category')
hour_df['year'] = hour_df.year.astype('category')
hour_df['month'] = hour_df.month.astype('category')
hour_df['holiday'] = hour_df.holiday.astype('category')
hour_df['weekday'] = hour_df.weekday.astype('category')
hour_df['workingday'] = hour_df.workingday.astype('category')
hour_df['weather_condition'] = hour_df.weather_condition.astype('category')

hour_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh cuaca terhadap jumlah pengguna sepeda?

```python
# Menggabungkan data berdasarkan musim
weather_agg = main_df.groupby("weather_condition_x").agg({
    "instant_x": "nunique",
    "count_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = weather_agg.index
y_max = weather_agg[('count_x', 'max')]
y_min = weather_agg[('count_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

# Modify x-axis labels
weather_labels = ['Clear to partly cloudy', 'Light snow to light rain', 'Misty cloudy to misty']
plt.xticks(x, weather_labels)

plt.xlabel('Weather Condition')
plt.ylabel('Number of Bike Sharing')
plt.title('Number of Maximum & Minimum Bike Sharing by Weather Condition')
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')
    
plt.grid(True)
plt.tight_layout()
plt.show()
```

Berdasarkan hasil di atas, fall season atau musim gugur memiliki jumlah penyewaan tertinggi.
